<h4>Combining all the mental health related questions group wise. For example, all ADHD questions into one column, all Anxiety questions into one column and so on

```python
ADHD = ['ADHD Q1', 'ADHD Q2', 'ADHD Q3', 'ADHD Q4']
df['ADHD'] = df[ADHD].sum(axis=1)

Anxiety = ['Anxiety Q1', 'Anxiety Q2']
df['Anxiety'] = df[Anxiety].sum(axis=1)

SelfEsteem = ['Self Esteem Q1', 'Self Esteem Q2','Self Esteem Q3']
df['Self Esteem'] = df[SelfEsteem].sum(axis=1)

Depression = ['Depression Q1', 'Depression Q2','Depression Q3']
df['Depression'] = df[Depression].sum(axis=1)

Total = ['ADHD', 'Anxiety','Self Esteem','Depression']
df['Total Score'] = df[Total].sum(axis=1)

df.drop(df.iloc[:, 7:19], inplace = True, axis = 1)
```

 >> Above statistical representation where we have taken the mean Anxiety Scores, show that excessive usage of social media may cause symptoms of Anxiety
